### Types of Statistics

1. Descriptive Statistics
2. Inferencial Statistics

**Descriptive Statistics**
Descriptive statistics are used to summarize and describe the main features of a dataset. They provide a clear and concise representation of the data's characteristics

 **Inferential statistics**
 Inferential statistics are used to draw conclusions or make inferences about a population based on a sample of data. They involve making predictions, testing hypotheses, and estimating parameters.

In Descriptive Statistics Datatypes are
1. Quantitative(Numerical)
2. Qualitative(Categorical)


**Quantitative(Numerical)** :

`Interval Data (Numeric)`:

Represents data with a consistent interval between values.
Has no true zero point.
Examples: Temperature in Celsius or Fahrenheit.

`Ratio Data (Numeric)`:

Represents data with a consistent interval between values.
Has a true zero point, which means that zero indicates the absence of the quantity being measured.
Examples: Height, weight, income, age.

**Qualitative(Categorical)** :

`Nominal Data (Categorical)`:

Represents categories or labels with no inherent order.
Examples: Colors, gender, types of fruits.

`Ordinal Data (Categorical)`:

Represents categories with a meaningful order or ranking.
Examples: Educational levels (e.g., high school, college, postgraduate), customer satisfaction ratings (e.g., very dissatisfied, dissatisfied, neutral, satisfied, very satisfied).

### Measures of Central Tendency:

**Mean**: The average of a set of values, calculated by summing all values and dividing by the number of values. 

It's represented as μ (mu).

**Median**: The middle value when data is sorted in ascending order.

It's less affected by extreme values and is a measure of the central point in the data.

**Mode**: The most frequently occurring value in the dataset.

### Measures of Spread:

**Range**: The "range" in statistics refers to the difference between the maximum and minimum values in a dataset.

**Variance**: A measure of how data points are spread out from the mean. It's the average of the squared differences between each data point and the mean.

**Standard Deviation**: The square root of the variance. It provides a more interpretable measure of the spread of the data.
Range: The difference between the maximum and minimum values in the dataset.

**Mean Absolute Deviation(MAD)** :It is a measure of the average absolute differences between data points and a central measure, typically the mean. MAD is used to quantify the overall variability or dispersion of a dataset. It is particularly useful when you want to understand how much, on average, individual data points differ from a central value.

**Interquartile Range (IQR)**: The range between the 25th and 75th percentiles, which measures the spread of the middle 50% of the data.

**Percentiles**: 
1. Suppose there is 100 data points are divided into 5 parts then it is called `Quaintile` it means 20% 

2. Suppose there is 100 data points are divided into 10 parts then it is called `Decile` it means 10%

3. Suppose there is 100 data points are divided into 4 parts then it is called `Quartile` it means 25%

### Variance

#### Variance (σ²) = Σ [(xᵢ - μ)²] / N
Where:

    σ² represents the variance.
    xᵢ represents each individual data point in the dataset.
    μ (mu) is the mean (average) of the dataset.
    N is the total number of data points.
    
   

### Standard Deviation

Standard Deviation (σ) = √Variance (σ²)

### Mean Absolute Deviation(MAD)

#### MAD = Σ |xᵢ - μ| / N

Where:

    Σ denotes the summation (sum of all absolute differences).
    xᵢ represents each individual data point in the dataset.
    μ (mu) is the mean (average) of the dataset.
    N is the total number of data points in the dataset.

# Measure of Relationship

### Covariance

 Covariance indicates whether an increase in one variable is associated with an increase or decrease in another variable.
 
 **cov(X,Y) = Σ(Xᵢ - μ)(Yᵢ - μ)/n**
 
 where,
 
          cov(X,Y) -- is the covariance between variables X and Y.
          n -- the number of data points.
          Xᵢ and Yᵢ are individual data points for variables X and Y.
          μ -- mean 

### Pearson Correlation Coefficient

It is a statistical measure that quantifies the strength and direction of the linear relationship between two continuous variables. 

The Pearson correlation coefficient can range from -1 to 1,

    If p = 1, it indicates a perfect positive linear relationship, meaning that as one variable increases, the other also       increases in a linear fashion.
    If p = -1, it indicates a perfect negative linear relationship, meaning that as one variable increases, the other           decreases in a linear fashion.
    If p = 0, it indicates no linear relationship between the two variables.
    
   **p = cov(x,y)/σ(x)σ(y)**
    
    where, 
            p  is the Pearson correlation coefficient.
           

### z-score

A z-score (also known as a standard score) is a statistical measure that quantifies how many standard deviations a data point is away from the mean of a distribution. It allows you to standardize data and compare values from different distributions by transforming them into a common scale.

The range of typical z-scores is usually between -3 and 3.

   **z= x−μ/σ**
   
                where,
                        z is the z-score.
                        x is the individual data point.
                        μ is the mean (average) of the dataset.
                        σ is the standard deviation of the dataset.




### Empirical Rule (68-95-99.7 Rule) :
According to the empirical rule, in a normal distribution, approximately 68% of the data falls within one standard deviation of the mean, 95% falls within two standard deviations, and 99.7% falls within three standard deviations. This means that the vast majority of data points have z-scores between -3 and 3 in a normal distribution.



### Outlier Detection: 
In many statistical applications, z-scores are used to identify outliers. Common thresholds for defining outliers include z-scores greater than 2 or less than -2, or sometimes z-scores greater than 3 or less than -3. Data points with z-scores outside these thresholds are considered potential outliers.
